```python
import torch
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n-seg.pt')
#model = YOLO('runs/detect/train2/weights/best.pt')
model.train(data='/content/drive/MyDrive/nagasaki/obb.yaml', epochs=1000, imgsz=640, device=[0], batch=16, patience=75, project='/content/drive/MyDrive/nagasaki/logs')
```

```python
model = YOLO('/content/drive/MyDrive/nagasaki/logs/train/weights/best.pt')
print(model.val())
```
